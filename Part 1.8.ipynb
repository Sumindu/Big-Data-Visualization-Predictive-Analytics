```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "Desktop/big data visualization/dataset/cleaned_transformed_dataset3.csv"
df = pd.read_csv(file_path)

# Display the columns to the user
print(df.columns)

# Prompt user to choose a column
chosen_column = 'clouds_all'

# Check if the column exists
if chosen_column in df.columns:
    # Histogram plot
    plt.figure(figsize=(10, 6))
    plt.hist(df[chosen_column].dropna(), bins=20, color="skyblue", edgecolor="black")
    plt.title(f"Histogram of {chosen_column}", fontsize=16)
    plt.xlabel(chosen_column, fontsize=12)
    plt.ylabel("Frequency", fontsize=12)
    plt.grid(axis="y", alpha=0.75)
    plt.show()

    # Scatter plot with a default comparison column (e.g., "traffic_volume")
    comparison_column = "traffic_volume" if "traffic_volume" in df.columns else input(
        "Enter the column for the Y-axis comparison: "
    )

    if comparison_column in df.columns:
        plt.figure(figsize=(10, 6))
        plt.scatter(df[chosen_column], df[comparison_column], alpha=0.6, color="coral")
        plt.title(f"Scatter Plot: {chosen_column} vs {comparison_column}", fontsize=16)
        plt.xlabel(chosen_column, fontsize=12)
        plt.ylabel(comparison_column, fontsize=12)
        plt.grid(alpha=0.5)
        plt.show()
    else:
        print(f"Comparison column '{comparison_column}' not found.")
else:
    print(f"Column '{chosen_column}' not found in the dataset.")
```
