```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
# file_path = "/mnt/data/cleaned_transformed_dataset1.csv"
file_path = "Desktop/big data visualization/dataset/cleaned_transformed_dataset3.csv"

df = pd.read_csv(file_path)

# Convert 'date_time' column to datetime
df['date_time'] = pd.to_datetime(df['date_time'], errors='coerce')

# Drop rows with invalid or missing dates
df = df.dropna(subset=['date_time'])

# Sort data by date_time for better visualization
df = df.sort_values('date_time')
    

# Scatter Plot
plt.figure(figsize=(12, 6))
plt.scatter(df['date_time'], df['traffic_volume'], alpha=0.6, color="coral", s=10)
plt.title("Traffic Volume Over Time", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Traffic Volume", fontsize=12)
plt.xticks(rotation=45)
plt.grid(alpha=0.5)
plt.tight_layout()
plt.show()

```
